```python
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
imgg = mpimg.imread("image.jpg")
plt.imshow(imgg, cmap = "gray")
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.show()
img = cv.imread("image.jpg",cv.IMREAD_COLOR)
cv.imshow("Original Image", img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# converting to grayscale
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv.imshow("Grayscale Image", gray)
plt.imshow(gray, cmap = "gray")
plt.title("Grayscale Image"), plt.xticks([]), plt.yticks([])
plt.show()
cv.waitKey(0)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# converting to hsv image
img_2 = mpimg.imread("image.jpg")
hsv = cv.cvtColor(img_2, cv.COLOR_BGR2HSV)
cv.imshow("HSV Image from CV", hsv)
plt.imshow(hsv, cmap = "gray")
plt.title("HSV Image"), plt.xticks([]), plt.yticks([])
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# edge detection
edges = cv.Canny(img, 100, 200)
plt.subplot(1, 2, 1)
cv.imshow("Edge Image", edges)
cv.imwrite('Edge Image.jpg', edges)
plt.imshow(imgg, cmap = "gray")
plt.title("Original Image"), plt.xticks([]), plt.yticks([])
plt.show()
plt.subplot(1, 2, 2) 
plt.imshow(edges, cmap = "gray")
plt.title("Edge Image"), plt.xticks([]), plt.yticks([])
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
# sobelx edge detection
sobelx = cv.Sobel(src=gray, ddepth=cv.CV_64F, dx=1, dy=0, ksize=5)
cv.imshow('Sobel X', sobelx)
plt.imshow(sobelx, cmap = "gray")
plt.title("SobelX Image"), plt.xticks([]), plt.yticks([])
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
image_street = cv.imread("group.jpg")
cv.imshow('Group', image_street)
plt.imshow(image_street, cmap = "Reds")
plt.title("Group Image"), plt.xticks([]), plt.yticks([])
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import imutils
hog = cv.HOGDescriptor()
hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())
image_street2 = imutils.resize(image_street, width=min(400, image_street.shape[1]))
(regions, _) = hog.detectMultiScale(image_street2, winStride=(4, 4), padding=(2, 2), scale=1.05)
for (x, y, w, h) in regions:
    cv.rectangle(image_street2, (x, y), (x + w, y + h), (255, 255, 255), 1)
cv.imshow("Detected Image", image_street2)
plt.imshow(image_street2, cmap = "Reds")
plt.title("Person Detected Image"), plt.xticks([]), plt.yticks([])
plt.show()
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
plt.title("Original"), plt.xticks([]), plt.yticks([])
plt.imshow(image_street)
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_image = cv.filter2D(image_street, -1, kernel)
cv.imwrite('sharpened_image.jpg', sharpened_image) 
plt.title("Sharpened Image"), plt.xticks([]), plt.yticks([])
plt.imshow(sharpened_image) 
plt.show()
```

```python
filtered_image2 = cv.GaussianBlur(image_street, (7, 7), 0)
cv.imwrite('Gaussian Blur.jpg', filtered_image2) 
plt.title("Gaussian Blur") 
plt.imshow(filtered_image2), plt.xticks([]), plt.yticks([])
plt.show()
```
