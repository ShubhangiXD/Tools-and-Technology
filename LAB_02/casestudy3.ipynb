```python
# initialising empty product dictionary
products = {}
cart = {}
checkout = 0
# initialising a class to store prodct information
class Products:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def __str__(self):
        return f"Name: {self.name}, Price: Rs{self.price}, Quantity: {self.quantity}"
        
# Function to add items to INVENTORY
def addInventory():
    name = input("Enter the name of the product: ")
    price = int(input("Enter the price of the product: "))
    quantity = int(input("Enter the quantity of the product: "))
    products[len(products)+1] = Products(name, price, quantity)
    print("Product Added:")
    print(products[len(products)])

# Function to view INVENTORY
def viewInventory(): 
    for key in products.keys():
        print("Product ID:", key, "\nProduct Information: ", products[key])

# Function to remove items from INVENTORY
def removeInventory():
    id = int(input("Enter the product ID to remove: "))
    x = products[id]
    print('Removing {} from the inventory.'.format(x.name))
    products.pop(id)
    if len(products)==0:
        print("No more items in inventory.")
    else:
        print("Modified Inventory: ")
        viewInventory()

def checkKey(dict, key):
    if key in dict.keys():
        return 1
    else:
        return 0
    
def modifyInventory(id, quantity):
    if quantity>products[id].quantity:
        return 0
    elif quantity == products[id].quantity:
        products.pop(id)
        return 1
    else:
        products[id].quantity = products[id].quantity - quantity
        return 1
    
def addUserCart():
    buy_id = int(input("Enter the product ID of the item you want to purchase: "))
    if checkKey(products, buy_id) == 0:
        print("The item is not available.")
    else:
        name = products[buy_id].name
        quantity = int(input("Enter quantity"))
        total = quantity * products[buy_id].price
        if modifyInventory(buy_id, quantity)== 0:
            print("Not enough items.")
        else:
            cart[buy_id] = Products(name, total, quantity)
        
def viewUserCart():
    print("Items in cart")
    for key in cart.keys():
        print("Product ID:", key, "\nProduct Information: ", cart[key])
        
def removeUserCart():
    id = int(input("Enter the product ID to remove: "))
    x = cart[id]
    print('Removing {} from the cart.'.format(x.name))
    products.pop(id)
    if len(products)==0:
        print("No more items in cart.")
    else:
        print("Modified cart: ")
        viewInventory()
        
def discounts():
    id = int(input("Enter the product ID on which the discount is to be applied: "))
    d = int(input("Enter discount amount"))
    products[id].price = products[id].price - (d/100)*products[id].price
    cart[id].price = products[id].price
    
def calCheckout():
    for key in cart:
        global checkout
        checkout += cart[key].price
    return checkout
    
            
def main():
    print("Add your first product in the inventory: ")
    addInventory()
    cont = True
    while cont == True:
        choices = {1: "Add to Inventory", 
                   2: "Remove from Inventory", 
                   3: "View Inventory", 
                   4: "Checkout", 
                   5: "Add to cart", 
                   6: "View User Cart", 
                   7: "Remove from User Cart",
                   8: "Discounts"}
        print(choices)
        choice = int(input("Enter your choice from the above: "))
        if choice == 1:
            addInventory()
        elif choice == 2:
            removeInventory(products)
        elif choice == 3:
            viewInventory()
        elif choice == 4:
            print("Your total checkout value is Rs{}.".format(calCheckout()))
            print("Thank you, exiting.")
            cont = False
        elif choice == 5:
            addUserCart()
        elif choice == 6:
            viewUserCart()
        elif choice == 7:
            removeUserCart()
        elif choice == 8:
            discounts()
        else:
            print("Wrong choice enter again: ")
        
main()

    
```
