```python
import pandas as pd
import numpy as np
crime=pd.read_csv('india-crime.csv')
print(crime)
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
maharashtra = crime[crime['STATE/UT']=="MAHARASHTRA"]
total = maharashtra['Kidnapping and Abduction'].sum()
print(total)
```

```python
raigad = maharashtra[maharashtra['DISTRICT'] == "RAIGAD"]
Total2 = raigad['Kidnapping and Abduction'].sum()
print(Total2)
```

```python
india_crime = crime[['Rape','Kidnapping and Abduction','Dowry Deaths','Assault on women with intent to outrage her modesty','Insult to modesty of Women','Cruelty by Husband or his Relatives','Importation of Girls']].copy()
corelationmatrix = india_crime.corr()
sb.heatmap(corelationmatrix,annot = True,cmap='coolwarm',fmt='.2f')
plt.title("Heatmap based on corelation matrix of 7 different columns in dataset.")
plt.show()
```

```python
filtered_data = crime[(crime['Year']>=2001) & (crime['Year']<=2012)]
filtered_data
plt.bar(filtered_data['Year'],filtered_data['Dowry Deaths'],color='skyblue')
plt.xlabel('Year')
plt.ylabel('Dowry Deaths')
plt.title('Dowry Deaths from 2001 to 2012')
plt.xticks(filtered_data['Year'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
filtered_data2 = crime[crime['STATE/UT'] == "WEST BENGAL"]
filtered_data3 = filtered_data2[(filtered_data2['Year']>=2001) & (filtered_data2['Year']<=2012)]
plt.bar(filtered_data3['Year'],filtered_data3['Dowry Deaths'],color='skyblue')
plt.xlabel('Year')
plt.ylabel('Dowry Deaths')
plt.title('Dowry Deaths from 2001 to 2012')
plt.xticks(filtered_data3['Year'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
